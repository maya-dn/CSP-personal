### 💎 Diamond Hack 2: String Algorithm Visualizer

**Challenge:** Create visualizations for string algorithms:
- Animated pattern matching (KMP, Boyer-Moore)
- String alignment visualization
- Compression algorithm demonstration
- Interactive algorithm comparison tool

**Tools you can use:**
- matplotlib for static plots
- HTML/CSS/JavaScript for interactive demos
- ASCII art for text-based visualizations
- Step-by-step algorithm walkthroughs

```python
# 💎 Diamond Hack 2: String Algorithm Visualizer

class StringAlgorithmVisualizer:
    """
    Visualize string algorithms step-by-step for educational purposes.
    """
    
    def visualize_pattern_search(self, text, pattern):
        """
        Visualize naive pattern searching algorithm step by step.
        
        Args:
            text (str): Text to search in
            pattern (str): Pattern to find
        """
        print(f"🔍 Searching for pattern '{pattern}' in text '{text}'")
        print("=" * 50)
        
        for i in range(len(text) - len(pattern) + 1):
            # Show current position
            print(f"\nStep {i+1}: Checking position {i}")
            
            # Visual representation
            text_display = list(text)
            pattern_display = [' '] * len(text)
            
            # Mark the current comparison window
            for j in range(len(pattern)):
                if i + j < len(text):
                    pattern_display[i + j] = pattern[j]
                    if text[i + j] == pattern[j]:
                        text_display[i + j] = f"[{text[i + j]}]"
                    else:
                        text_display[i + j] = f"({text[i + j]})"
                        break
            
            print("Text:   ", ''.join(text_display))
            print("Pattern:", ''.join(pattern_display))
            
            # Check if pattern matches at current position
            if text[i:i+len(pattern)] == pattern:
                print("✅ MATCH FOUND!")
                return i
            else:
                print("❌ No match, continue...")
        
        print("\n🚫 Pattern not found in text")
        return -1
    
    def visualize_string_comparison(self, str1, str2):
        """
        Visualize character-by-character string comparison.
        
        Args:
            str1 (str): First string
            str2 (str): Second string
        """
        print(f"🔄 Comparing '{str1}' with '{str2}'")
        print("=" * 40)
        
        max_len = max(len(str1), len(str2))
        matches = 0
        
        for i in range(max_len):
            char1 = str1[i] if i < len(str1) else '∅'
            char2 = str2[i] if i < len(str2) else '∅'
            
            match_indicator = "✅" if char1 == char2 and char1 != '∅' else "❌"
            if char1 == char2 and char1 != '∅':
                matches += 1
            
            print(f"Position {i}: '{char1}' vs '{char2}' {match_indicator}")
        
        similarity = matches / max_len * 100 if max_len > 0 else 0
        print(f"\n📊 Similarity: {similarity:.1f}% ({matches}/{max_len} matches)")

# Example usage
visualizer = StringAlgorithmVisualizer()

print("🎨 String Algorithm Visualizer Demo")
print("\n" + "="*60)

# Demonstrate pattern search visualization
visualizer.visualize_pattern_search("HELLO WORLD", "LLO")

print("\n" + "="*60)

# Demonstrate string comparison visualization
visualizer.visualize_string_comparison("PYTHON", "PISTON")
```
