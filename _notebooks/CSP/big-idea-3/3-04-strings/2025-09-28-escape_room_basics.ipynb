### I Do: Demo (AP Pseudocode)
```text
// Given s ← "COMPUTER SCIENCE"
LENGTH(s) // 17
CONCAT("AP ", "CSP") // "AP CSP"
subString(s, 1, 8) // "COMPUTER"
// Traverse
i ← 1
REPEAT UNTIL i > LENGTH(s)
  DISPLAY subString(s, i, i)
  i ← i + 1
```

### I Do: Demo (Python)
```python
s = "COMPUTER SCIENCE"
len(s)             # 17
"AP " + "CSP"     # "AP CSP"
s[0:8]             # "COMPUTER"  (Python is 0-based, end-exclusive)
for ch in s:
    print(ch)
```

````python
# AP CSP Required Task 1: first_n
# In AP pseudocode: subString(s, 1, n)
# In Python: s[0:n]
def first_n(s: str, n: int) -> str:
    return s[:max(0, n)]

# AP CSP Required Task 2: join_name
# CONCAT(first, " ", last)
def join_name(first: str, last: str) -> str:
    return (first or "").strip() + (" " if first and last else "") + (last or "").strip()

# AP CSP Required Task 3: count_vowels (simple traversal)
def count_vowels(s: str) -> int:
    vowels = set("aeiouAEIOU")
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
    return count

# Quick checks
assert first_n("COMPUTER", 4) == "COMP"
assert join_name("AP", "CSP") == "AP CSP"
assert count_vowels("ALGORITHM") == 3  # A O I
print("✅ AP CSP required tasks: basic checks passed")
````

    # Clean and normalize
    text = messy_text.strip()
    if not text:  # Handle empty string
        return ""
    
    # Normalize spaces and convert to title case
    text = ' '.join(text.split()).title()
    
    # Add period if needed, but avoid double periods
    if not text.endswith('.'):
        text += '.'
    
    return text
